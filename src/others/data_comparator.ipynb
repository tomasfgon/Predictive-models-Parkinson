```python
import pandas as pd
import matplotlib.pyplot as plt
import os

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python


csv_files = [
    '../data/tables/italian_original_features.csv',
    '../data/tables/mdvr_original_features.csv',
    '../data/tables/ah_original_features.csv'
]

# Initialize DataFrames
italian_df = pd.read_csv(csv_files[0])
mdvr_df = pd.read_csv(csv_files[1])
ah_df = pd.read_csv(csv_files[2])

# Remove the "Duration" column
italian_df = italian_df.drop("Duration", axis=1)
mdvr_df = mdvr_df.drop("Duration", axis=1)
ah_df = ah_df.drop("Duration", axis=1)

# Split DataFrames based on the "PD" column
italian_pd_0 = italian_df[italian_df["PD"] == 0].drop("PD", axis=1)
italian_pd_1 = italian_df[italian_df["PD"] == 1].drop("PD", axis=1)

mdvr_pd_0 = mdvr_df[mdvr_df["PD"] == 0].drop("PD", axis=1)
mdvr_pd_1 = mdvr_df[mdvr_df["PD"] == 1].drop("PD", axis=1)

ah_pd_0 = ah_df[ah_df["PD"] == 0].drop("PD", axis=1)
ah_pd_1 = ah_df[ah_df["PD"] == 1].drop("PD", axis=1)
```

```python
# Replace 'path_to_csv_files' with the actual path where your CSV files are located
csv_files = [
    'audio_teste1.wav_reduced_features.csv',
    'audio_teste2.wav_reduced_features.csv',
    'audio_teste3.wav_reduced_features.csv',
    'audio_teste4.wav_reduced_features.csv',
    'audio_teste5.wav_reduced_features.csv',
    'audio_teste6.wav_reduced_features.csv',
    'audio_teste7.wav_reduced_features.csv',
    'audio_teste8.wav_reduced_features.csv',
    'audio_teste9.wav_reduced_features.csv',
    'audio_teste10.wav_reduced_features.csv',
    'audio_teste11.wav_reduced_features.csv',
    'audio_teste12.wav_reduced_features.csv',
    'audio_teste13.wav_reduced_features.csv',

]

# Initialize an empty DataFrame
combined_df_testes = pd.DataFrame()

# Loop through each CSV file and load it into the DataFrame
for file in csv_files:
    file_path = os.path.join('../data/tables/testfiles', file)
    df = pd.read_csv(file_path)
    combined_df_testes = pd.concat([combined_df_testes, df], ignore_index=True)

# Display the combined DataFrame
combined_df_testes
```

```python

# Calculate the average values for each column
combined_avg = combined_df_testes.mean()
ah_pd_0_avg = ah_pd_0.mean()
ah_pd_1_avg = ah_pd_1.mean()
italian_pd_0_avg = italian_pd_0.mean()
italian_pd_1_avg = italian_pd_1.mean()
mdvr_pd_0_avg = mdvr_pd_0.mean()
mdvr_pd_1_avg = mdvr_pd_1.mean()

# Create a new DataFrame to store the mean values
mean_comparison_df = pd.DataFrame({
    'Variable': italian_pd_0_avg.index,
    'Test Files': combined_avg.values,
    'AH PD': ah_pd_1_avg.values,
    'AH HC': ah_pd_0_avg.values,
    'Italian PD': italian_pd_1_avg.values,
    'Italian HC': italian_pd_0_avg.values,
    'MDVR PD': mdvr_pd_1_avg.values,
    'MDVR HC': mdvr_pd_0_avg.values
})

# Display the mean values comparison table
(mean_comparison_df)
```

```python
# Number of subplots (equal to the number of columns)
num_columns = len(combined_df_testes.columns)

# Number of columns to display in a single row
num_columns_per_row = 3

# Calculate the number of rows needed
num_rows = (num_columns + num_columns_per_row - 1) // num_columns_per_row

# Plotting
fig, axes = plt.subplots(nrows=num_rows, ncols=num_columns_per_row, figsize=(15, 3 * num_rows), sharex=False)

# Flatten the 2D array of subplots to simplify indexing
axes = axes.flatten()

# Plot each column on a separate subplot
for i, column in enumerate(combined_df_testes.columns):
    axes[i].bar(['Custom', 'AH-PD','AH_HC','I-PD', 'I-HC', 'M-PD', 'M-HC'], [combined_avg[column], ah_pd_1_avg[column], ah_pd_0_avg[column], italian_pd_1_avg[column], italian_pd_0_avg[column], mdvr_pd_1_avg[column], mdvr_pd_0_avg[column]])
    axes[i].set_ylabel(column)

# Hide any empty subplots
for i in range(num_columns, num_columns_per_row * num_rows):
    fig.delaxes(axes[i])

# Set plot properties
plt.suptitle('Comparison of Average Values for Each Column', y=1.02)
plt.xlabel('Dataframes')

# Adjust layout to avoid label cut-off
plt.subplots_adjust(top=0.95, hspace=0.4, wspace=0.4)  # Adjust top and wspace values

# Show the plot
plt.show()
```
